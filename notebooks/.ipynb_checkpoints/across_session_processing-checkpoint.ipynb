```python
def minimum_state_duration(state_sequence, min_duration: int):
    """ Remove states that are too short"""

    # Apply the minimum duration constraint to the state sequence
    filtered_state_sequence = []
    current_state = state_sequence[0]
    current_duration = 1

    for i in range(1, len(state_sequence)):
        if state_sequence[i] == current_state:
            current_duration += 1
        else:
            if current_duration >= min_duration:
                filtered_state_sequence.extend([current_state] * current_duration)
            else:
                filtered_state_sequence.extend([state_sequence[i-1]] * current_duration)
            current_state = state_sequence[i]
            current_duration = 1

    # Extend the state sequence for the remaining time points
    if current_duration >= min_duration:
        filtered_state_sequence.extend([current_state] * current_duration)
    else:
        filtered_state_sequence.extend([state_sequence[-1]] * current_duration)
        
    return filtered_state_sequence
```

```python
import pandas as pd

# Dummy DataFrame
data = {
    'timestamp': pd.date_range(start='2023-06-01', periods=100, freq='D'),
    'activity_duration': [i for i in range(100)],
    'sport': ['running' if i % 2 == 0 else 'swimming' for i in range(100)]
}

df = pd.DataFrame(data)

# Convert 'timestamp' column to pandas datetime, if not already
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Set 'timestamp' as the index, this will make rolling operations more straightforward
df.set_index('timestamp', inplace=True)

# Sort the DataFrame by the index
df.sort_index(inplace=True)
# df.reset_index(inplace=True)

# Initialize an empty DataFrame to store the rolling averages
rolling_avg_df = pd.DataFrame()

# Loop through each unique day and each unique sport
for day in pd.date_range(start=df.timestamp.min(), end=df.timestamp.max(), freq='D'):

    day_data = df.loc[df.timestamp == day]
    print(day_data)
    for sport in day_data['sport'].unique():
        sport_data = day_data[day_data['sport'] == sport]
        
        if sport_data.empty:
            continue
        
        # Calculate the 7-day rolling average for this sport
        if len(sport_data)>6:
            rolling_avg = sport_data['activity_duration'].rolling(window='7D').mean()

            # Create a temporary DataFrame to store this data
            temp_df = pd.DataFrame({
                'timestamp': sport_data.index,
                '7_day_rolling_avg': rolling_avg,
                'sport': sport
            })

            # Append the temporary DataFrame to our results DataFrame
            rolling_avg_df = pd.concat([rolling_avg_df, temp_df])

# Reset the index for the final DataFrame
rolling_avg_df.reset_index(drop=True, inplace=True)

print(rolling_avg_df)
```

# calculate fatigue

```python
def get_fatigue(df_all_fits):
    # Convert the start_time column to datetime
    df_sel = df_all_fits.loc[df_all_fits.trigger == 'activity_end'].dropna(axis=1,how='all') 
    df = df_sel[['sport','start_time','total_timer_time']]
    df['start_time'] = pd.to_datetime(df['start_time'])

    # Convert the total_timer_time from seconds to minutes
    df['total_timer_time'] = df['total_timer_time'] / 60

    # Set the date as the index and drop the time component
    df.set_index('start_time', inplace=True)

    # Resample the DataFrame to an hourly frequency
    df_hourly = df.copy().resample('H').sum()

    # Initialize fatigue at 0
    df_hourly['fatigue'] = 0

    # Calculate fatigue
    fatigue = 0
    for i, row in df_hourly.iterrows():
        # Decrease fatigue by 1 for each hour that passes
        fatigue -= 1

        # Ensure fatigue doesn't go below 0
        fatigue = max(fatigue, 0)

        # Increase fatigue by the duration of the exercise session times 10
        fatigue += row['total_timer_time'] * 0.4

        # Update the fatigue score in the DataFrame
        df_hourly.at[i, 'fatigue'] = fatigue

    # Check the DataFrame
    return df_hourly 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

# Load the data
df_sel = df_all_fits.loc[df_all_fits.trigger == 'activity_end'].dropna(axis=1,how='all') 
df = df_sel[['sport','start_time','total_timer_time']]

# Convert the start_time column to datetime
df['start_time'] = pd.to_datetime(df['start_time'])

# Convert the total_timer_time from seconds to minutes
df.loc[:,'total_timer_time'] = df['total_timer_time'] / 60

# Set the date as the index and drop the time component
df.set_index('start_time', inplace=True)
df.index = df.index.date

# Reshape the dataframe to get total time per sport per day
df_reshaped = df.pivot_table(values='total_timer_time', index=df.index, columns='sport', aggfunc=np.sum)

# Reindex to include all dates from the earliest to the last (plus 1 week)
new_index = pd.date_range(start=df_reshaped.index.min(), end=df_reshaped.index.max() + pd.DateOffset(weeks=1))
df_reshaped = df_reshaped.reindex(new_index, fill_value=0)

# Reset index to make date a column again for plotting
df_reshaped.reset_index(inplace=True)
df_reshaped.rename(columns={'index': 'date'}, inplace=True)

# Melt the dataframe to prepare it for scatter plot
df_melted = df_reshaped.melt(id_vars='date', var_name='sport', value_name='total_time')

# Filter out rows with total_time equal to 0
df_melted = df_melted[df_melted['total_time'] != 0]

# Set style
sns.set(style="whitegrid")

fig, (ax1, ax2) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [2, 1]}, sharex=True, figsize=(6,9))


# Plotting
relplot = sns.scatterplot(
    data=df_melted,
    x="date", 
    y="sport", 
    hue="sport",
    size="total_time",
    palette="tab10",
    sizes=(10, 200),
#     height=8,
#     aspect=1.5,
    ax=ax2
)

ax2.set_ylim([-1,6])
# # Create a second y-axis for fatigue
# ax2 = ax1.twinx()
# ax2.plot(df.index, df_hourly['fatigue'], color='black', linewidth=1, linestyle='dashed')


# Adjust legend
# Add a legend to the first subplot only
# legend = ax2.legend()  # Get the legend object
# legend.set_bbox_to_anchor((1.05, 0.6))  # Adjust the legend position
ax2.legend().remove()

# Set x-tick for every Monday
# ax2.xaxis.set_minor_locator(mdates.WeekdayLocator(byweekday=mdates.MO))
# ax2.xaxis.set_minor_formatter(mdates.DateFormatter('%d'))


# Set x-tick with a label for the first day of a month
relplot.xaxis.set_major_locator(mdates.MonthLocator())
relplot.xaxis.set_major_formatter(mdates.DateFormatter('%b\n%Y'))


df_hourly = get_fatigue(df_all_fits)

ax1.plot(df_hourly.index, df_hourly['fatigue'], color='black', linewidth=1, linestyle='dashed')
```

```python
import random

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.collections import LineCollection
import numpy as np
import datetime

# Extract required columns 
df = df_all_fits.loc[df_all_fits.trigger == 'activity_end'].dropna(axis=1,how='all')

# Convert 'start_time' to time of day
df['start_time_of_day'] = (pd.to_datetime(df['start_time']) + pd.to_timedelta(3600, unit='s')).dt.time

# Convert 'total_timer_time' to timedelta  
df.loc[:,'duration'] = pd.to_timedelta(df['total_timer_time'], unit='s')

# Calculate 'end_time' and time of day
df.loc[:,'end_time_of_day'] = (pd.to_datetime(df['start_time']) + df['duration']  + pd.to_timedelta(3600, unit='s')).dt.time  
# df.loc[:,'end_time_of_day'] = df['end_time'].dt.time

# Convert times to datetime for plotting
df['start_time_of_day'] = df['start_time_of_day'].apply(lambda x: datetime.datetime.combine(datetime.date.today(), x))
df['end_time_of_day'] = df['end_time_of_day'].apply(lambda x: datetime.datetime.combine(datetime.date.today(), x))

# Group by 'sport' for different colors
groups = df.groupby('sport')


# Create a color map
colors = plt.cm.rainbow(np.linspace(0, 1, len(groups)))


fig, ax = plt.subplots(figsize=(12, 8))

# Assign each sport a unique index on the x-axis
sports = df['sport'].unique()
sport_indices = {sport: i for i, sport in enumerate(sports)}

# Create a color map
colors = plt.cm.rainbow(np.linspace(0, 1, len(sports)))

for sport, color in zip(sports, colors):
    sport_data = df[df['sport'] == sport]
    x = [sport_indices[sport] + random.uniform(-0.2, 0.2) for _ in range(len(sport_data))]  # add random offset to x-position
    ax.scatter(x, mdates.date2num(sport_data['start_time_of_day']), color=color, label=sport)
    lc = LineCollection([list(zip([xi, xi], mdates.date2num([start, end]))) for xi, start, end in zip(x, sport_data['start_time_of_day'], sport_data['end_time_of_day'])], colors=color)
    ax.add_collection(lc)

# Format y-axis to display time
ax.yaxis_date()
ax.yaxis.set_major_formatter(mdates.DateFormatter('%H:%M:%S'))

plt.xticks(range(len(sports)), sports)
plt.xlabel('Sport')
plt.ylabel('Time of Day')
plt.title('Activity Start Times and Durations by Sport')
plt.legend()
plt.tight_layout()
plt.show()
```
