```python
import numpy as np
import pandas as pd
import os
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data_april_may.csv')
df["Date"] = pd.to_datetime(df["Date"])
df['day'] = df['Date'].dt.date
df["Days_to_Current_Date"] = (datetime.now() - df["Date"]).dt.days

# select days
df_month = df.loc[df['Days_to_Current_Date'] < 31,:]

# Distance to float
df_month['Distance'] = df_month['Distance'].str.replace(',', '.')
df_month.Distance = df_month.Distance.apply(float)
```

```python


# Convert the 'Elapsed_time' column to timedelta
df_month['Elapsed Time'] = pd.to_timedelta(df_month['Elapsed Time'])

# Convert timedelta to minutes
df_month['workout_minutes'] = np.round(df_month['Elapsed Time'].dt.total_seconds()/60)
```
